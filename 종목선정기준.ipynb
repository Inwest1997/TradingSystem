```python
ascend = []
for idx in range(len(goog)):
    if goog.iloc[idx]['MA(120)']< goog.iloc[idx]['MA(60)'] and goog.iloc[idx]['MA(60)']< goog.iloc[idx]['MA(20)'] and goog.iloc[idx]['MA(20)']< goog.iloc[idx]['MA(5)'] and goog.iloc[idx]['MA(120)']< goog.iloc[idx]['Adj Close']:
        ascend.append(1)
    else:        
        ascend.append(0)
```

```python
import numpy as np
```

```python
s1 = []
for idx in range(len(goog)):
    if idx < 20:
        s1.append(np.NAN)
    else:
        if goog.iloc[idx]['High']>goog.iloc[idx-20]['High']:
            s1.append(1)
        else:
            s1.append(0)
```

```python
goog['high20'] = s1
goog.dropna(inplace=True)
```

```python
s2 = []
for idx in range(len(goog)):
    if idx < 1:
        s2.append(np.NAN)
    else:
        if goog.iloc[idx]['Volume']>goog.iloc[idx-1]['Volume']*3:
            s2.append(1)
        else:
            s2.append(0)
```

```python
goog['volume200'] = s2
goog.dropna(inplace=True)
```

```python
s3 = []
for idx in range(len(goog)):
    if idx < 2:
        s3.append(np.NAN)
    else:
        if goog.iloc[idx-1]['Volume']>goog.iloc[idx-2]['Volume'] and goog.iloc[idx-1]['Adj Close']>goog.iloc[idx-2]['Adj Close'] :
            s3.append(1)
        else:
            s3.append(0)
```

```python
goog['dm'] = s3
goog.dropna(inplace=True)
```
